```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('data.csv')
df
```

```python
plt.xlabel('Year')
plt.ylabel('Per Capita($)')
plt.scatter(df.year, df.per_capita_income)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(df[['year']], df.per_capita_income)
reg.predict([[2017]])
```

```python
# y = mx + c
m = reg.coef_
c = reg.intercept_
y = m * 2018 + c
y
```

```python
da = pd.read_csv('data.csv')
da.head()
```

```python
per_capita = reg.predict(da[['year']])
```

```python
da['per_capita_income'] = per_capita
da.to_csv('prediction.csv')
```
